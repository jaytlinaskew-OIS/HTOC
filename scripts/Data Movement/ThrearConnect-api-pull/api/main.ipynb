```python
import requests
from requests.auth import HTTPBasicAuth
import urllib3
import sys
import os
import json

# Add your project root to sys.path
project_root = r"C:\Users\jaskew\Documents\project_repository\scripts\Data Movement\ThrearConnect-api-pull"
if project_root not in sys.path:
    sys.path.append(project_root)

sys.path.append(os.path.abspath(r"C:\Users\jaskew\Documents\project_repository\scripts\Data Movement\ThrearConnect-api-pull"))

from utils.config_loader import load_config

config_path = r"C:\Users\jaskew\Documents\project_repository\scripts\Data Movement\ThrearConnect-api-pull\utils\config.json"

api_key, access_id, url_base = load_config(config_path)

print("Loading config from:", config_path)
print("Base url:", url_base)

# Disable SSL verification (only for internal/testing use)
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
verify_ssl = False

# Set up the endpoint and parameters
endpoint = "/indicators"
url = f"{url_base}{endpoint}"
params = {
    "resultLimit": 10
}

print(url)
# Perform the request
response = requests.get(
    url,
    auth=HTTPBasicAuth(access_id, api_key),
    params=params,
    verify=verify_ssl
)

# Handle the response
if response.status_code == 200:
    print("Success:")
    print(json.dumps(response.json(), indent=2))
else:
    print("Error:", response.status_code)
    print(response.text)
```
