```python
"""
Load Observation data
 
"""
import pandas as pd
import datetime
import os
from dateutil.relativedelta import relativedelta
from datetime import datetime
from datetime import timedelta
from datetime import date
```

```python
# Define parameters
days_back = 7
start_full_date_str = "2024-01-02"
start_date_str = "2024-01-02"
# Define the file path template
file_path_template = "Z:/HTOC/Data_Analytics/Data/OpDiv_Observations/htoc_opdiv_obs_d{date}.csv"
# Define the save directory
save_directory = r"C:\Users\jaskew\Documents\project_repository\data\raw\ObservationDataFiles"
```

```python
# Calculate today's date
today = datetime.today()
print(today)

# Calculate end date (today + 0 days)
end_dt = today + timedelta(days=0)
end_date_str = end_dt.strftime("%Y-%m-%d")
print(end_date_str)

# Convert string dates to datetime objects
start_full_dt = datetime.strptime(start_full_date_str, "%Y-%m-%d")
start_dt = datetime.strptime(start_date_str, "%Y-%m-%d")
```

```python
# Function to load files and save them to the specified directory
def load_and_save_files(filenames, save_directory):
    #check if file dates exist in save directory
    existing_files = os.listdir(save_directory)
    for filename in filenames:
        file_date = filename.split('d')[1].split('.')[0]
        if file_date in existing_files:
            print(f"File {filename} already exists in {save_directory}. Skipping.")
            continue
        
    os.makedirs(save_directory, exist_ok=True)  # Ensure the directory exists
    for filename in filenames:
        if not os.path.exists(filename):
            print(f"File {filename} does not exist. Skipping.")
            continue
        df = pd.read_csv(filename)
        save_path = os.path.join(save_directory, os.path.basename(filename))
        df.to_csv(save_path, index=False)  # Save the file
        yield df

# Define the file path template and date range
date_format = "%Y%m%d"
start_date = datetime.strptime(start_date_str, "%Y-%m-%d")  # Convert start_date_str to datetime
dates_to_pull = pd.date_range(start_date, end_dt, freq='D')

# Generate the list of file paths
datelist = [file_path_template.format(date=dt.strftime(date_format)) for dt in dates_to_pull]
print(datelist)

# Concatenate dataframes from the list of files
src = pd.concat(load_and_save_files(datelist, save_directory))

# Clean up the 'indicator' and 'OpDiv' columns
src['indicator'] = src['indicator'].str.split(' ', expand=True)[0].str.strip()
src['OpDiv'] = src['OpDiv'].str.strip()
```
