```python
import timeit, tracemalloc, numpy as np

# ── 1. Inefficient: Python loop (object-by-object) ──
def inefficient_loop(n=1_000_000):
    total = 0
    for i in range(n):
        total += (i*i + 3*i) % 7     # every iteration creates 3-4 Python ints
    return total

# ── 2. Efficient: NumPy vectorised (bulk, in C) ──
def efficient_numpy(n=1_000_000):
    arr = np.arange(n, dtype=np.int64)           # 8 MB contiguous buffer
    res = (arr*arr + 3*arr) % 7                  # done by SIMD-optimised C code
    return res.sum()

# quick helper for timing + peak-RAM
def bench(fn):
    wall = timeit.timeit(lambda: fn(), number=3) / 3
    tracemalloc.start(); fn(); _, peak = tracemalloc.get_traced_memory(); tracemalloc.stop()
    return wall, peak/1_048_576             # MiB
```

```python
# Call inefficient_loop function
result_inefficient = inefficient_loop()
print(f"Result of inefficient_loop: {result_inefficient}")
```
