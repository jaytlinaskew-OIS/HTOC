## Tag icon library

**`entries`** is nested by category id: **`entries.phishing["Phishing for Information"]`** → `{ "icon_url": "..." }` so all phishing tags sit in the same JSON object. **`categories`** mirrors metadata: **`label`**, **`icon`**, **`includes`**, **`tags`**. **`icon_overrides`** forces a Lucide id per tag. Sync preserves category membership; new tags match `includes` or go to **`other`**. Use `tag_html_span` for previews (`get_entry_for_tag` resolves lookups in code).
