```python
# List all subfolders in folder_path
logs_folder = r'Z:\HTOC\Data_Analytics\Data\OpDiv_Predictions\Logs'

subfolders = [f.path for f in os.scandir(logs_folder) if f.is_dir()]
for subfolder in subfolders:
    print(f"Folder: {os.path.basename(subfolder)}")
    subfolder_files = [f for f in os.listdir(subfolder) if os.path.isfile(os.path.join(subfolder, f))]
    print(f"  Files: {subfolder_files}")
```

```python
excel_dfs = {}
for subfolder in subfolders:
    subfolder_files = [f for f in os.listdir(subfolder) if os.path.isfile(os.path.join(subfolder, f))]
    for file in subfolder_files:
        # Skip temporary Excel files (usually start with '~$')
        if file.endswith('.xlsx') and not file.startswith('~$'):
            file_full_path = os.path.join(subfolder, file)
            var_base = os.path.splitext(file)[0]
            excel_dfs[var_base] = pd.read_excel(file_full_path, engine='openpyxl')
            print(f"Loaded {file} from {os.path.basename(subfolder)} into excel_dfs['{var_base}'], shape: {excel_dfs[var_base].shape}")
```

```python
import os
import shutil
from datetime import datetime
from typing import Dict
import pandas as pd
from IPython.display import display

# ---------------------------------------------------------------------
# Normalize helper
CONFIDENCE_MAP = {
    '30-day: highly likely': 'Highly Likely',
    '30-day: possibly active': 'Possibly Active',
    '30-day: low confidence': 'Low Confidence',
    '14-day: highly likely': 'Highly Likely',
    '14-day: possibly active': 'Possibly Active',
    '14-day: low confidence': 'Low Confidence',
    '7-day: highly likely': 'Highly Likely',
    '7-day: possibly active': 'Possibly Active',
    '7-day: low confidence': 'Low Confidence',
    'highly likely': 'Highly Likely',
    'possibly active': 'Possibly Active',
    'low confidence': 'Low Confidence'
}

# ---------------------------------------------------------------------
# Normalize daily reports by FileDate

dfs_by_date: Dict[pd.Timestamp, pd.DataFrame] = {}
print("\n✅ Daily files successfully parsed by FileDate:")

for key, df in dfs.items():
    df.columns = df.columns.str.strip()
    if 'FileDate' not in df.columns:
        continue

    df = df.copy()
    if pd.api.types.is_datetime64_any_dtype(df['FileDate']):
        df['FileDate'] = pd.to_datetime(df['FileDate']).dt.normalize()
    else:
        df['FileDate'] = pd.to_datetime(
            df['FileDate'].astype(str).str.zfill(8),
            format='%Y%m%d', errors='coerce'
        ).dt.normalize()

    valid_dates = df['FileDate'].dropna().unique()
    if len(valid_dates) > 0:
        dfs[key] = df
        for dt in valid_dates:
            dfs_by_date[dt] = df
        print(f"- {key} → {[d.strftime('%Y-%m-%d') for d in valid_dates]}")

# ---------------------------------------------------------------------
# Backup and clean forecast logs

backup_folder = r'Z:\HTOC\Data_Analytics\Data\OpDiv_Predictions\Logs_Backup'
os.makedirs(backup_folder, exist_ok=True)
print("\n🔄 Backing up and cleaning existing forecast logs...")

for subfolder in subfolders:
    partner_name = os.path.basename(subfolder).strip()
    log_path = os.path.join(subfolder, "forecast_log.xlsx")
    if os.path.exists(log_path):
        backup_path = os.path.join(backup_folder, f"{partner_name}_forecast_log_backup_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx")
        shutil.copy2(log_path, backup_path)
        print(f"  Backed up {partner_name} → {backup_path}")
        os.remove(log_path)
        print(f"   Deleted original {partner_name} log")

print("\n" + "="*60)
print(" Starting fresh data processing with clean normalization...")
print("="*60)

# ---------------------------------------------------------------------
# Process forecast logs per partner

for subfolder in subfolders:
    partner_name = os.path.basename(subfolder).strip()
    partner_clean = partner_name.lower()
    log_key = f"{partner_name}_forecast_log"
    log_path = os.path.join(subfolder, "forecast_log.xlsx")

    if log_key not in excel_dfs:
        print(f" Skipping {partner_name}: No forecast log found.")
        continue

    log_df = excel_dfs[log_key].copy()
    log_df.columns = log_df.columns.str.strip()

    log_df['Forecasted Check Date'] = pd.to_datetime(log_df['Forecasted Check Date'], errors='coerce').dt.normalize()
    log_df['Outcome'] = log_df.get('Outcome', '')
    log_df['Indicator_norm'] = log_df['Indicator'].astype(str).str.strip().str.lower()

    # Normalize confidence
    log_df['Confidence'] = (
        log_df['Confidence']
        .astype(str)
        .str.strip()
        .str.lower()
        .map(CONFIDENCE_MAP)
        .fillna(log_df['Confidence'])
    )

    # Normalize forecast type
    log_df['Forecast Type'] = (
        log_df['Forecast Type']
        .astype(str)
        .str.split(':')
        .str[-1]
        .str.strip()
        .str.title()
    )

    for idx, row in log_df.iterrows():
        check_date = row['Forecasted Check Date']
        indicator_norm = row['Indicator_norm']
        daily_df = dfs_by_date.get(check_date)
        if daily_df is None:
            continue

        if 'Indicator_norm' not in daily_df.columns:
            daily_df['Indicator_norm'] = daily_df['Indicator'].astype(str).str.strip().str.lower()
        if 'Partner_norm' not in daily_df.columns:
            daily_df['Partner_norm'] = daily_df['Partner'].astype(str).str.strip().str.lower()

        subset = daily_df[
            (daily_df['Partner_norm'] == partner_clean) &
            (daily_df['Indicator_norm'] == indicator_norm)
        ]
        seen = subset['Observed Today'].eq(1).any()
        log_df.at[idx, 'Outcome'] = 'Seen' if seen else 'Not Seen'

    print(f"\n {partner_name} – Outcome counts:")
    print(log_df['Outcome'].value_counts(dropna=False).to_string())
    display(log_df[log_df['Outcome'] == 'Seen'].head(50))

    # Merge with existing file if exists
    if os.path.exists(log_path):
        try:
            existing_df = pd.read_excel(log_path)
            existing_df.columns = existing_df.columns.str.strip()
            existing_df['Confidence'] = (
                existing_df['Confidence']
                .astype(str)
                .str.strip()
                .str.lower()
                .map(CONFIDENCE_MAP)
                .fillna(existing_df['Confidence'])
            )
            existing_df['Forecast Type'] = (
                existing_df['Forecast Type']
                .astype(str)
                .str.split(':')
                .str[-1]
                .str.strip()
                .str.title()
            )
        except Exception as e:
            print(f" Could not read existing file for {partner_name}: {e}")
            existing_df = pd.DataFrame(columns=log_df.columns)
    else:
        existing_df = pd.DataFrame(columns=log_df.columns)

    combined_df = pd.concat([existing_df, log_df], ignore_index=True)
    combined_df['Forecast Date'] = pd.to_datetime(combined_df['Forecast Date'], errors='coerce')

    combined_df['Confidence'] = (
        combined_df['Confidence']
        .astype(str)
        .str.strip()
        .str.lower()
        .map(CONFIDENCE_MAP)
        .fillna(combined_df['Confidence'])
    )
    combined_df['Forecast Type'] = (
        combined_df['Forecast Type']
        .astype(str)
        .str.split(':')
        .str[-1]
        .str.strip()
        .str.title()
    )

    combined_df.drop_duplicates(
        subset=['Indicator', 'Forecast Date', 'Forecast Type', 'Confidence'],
        keep='last', inplace=True
    )

    combined_df.sort_values(by=['Indicator', 'Forecasted Check Date'], inplace=True)
    if 'Indicator_norm' in combined_df.columns:
        combined_df.drop(columns='Indicator_norm', inplace=True)

    try:
        combined_df.to_excel(log_path, index=False)
        print(f" Updated and saved: {log_path}")
    except Exception as e:
        print(f" Failed to save log for {partner_name}: {e}")
```
