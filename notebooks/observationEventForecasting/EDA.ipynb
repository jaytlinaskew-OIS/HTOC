```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/jaytlinaskew/GitRepository/TimeSeries-Analysis/notebooks/observationEventForecasting/DataPreprocessing/FullIndicatorMatrix.csv')
data.head()
```

```python
correct = data[data['indicator'] == '104.21.48.1']
correct
```

```python
# Ensure 'date' column in 'correct' DataFrame is in datetime format
correct.loc[:, 'date'] = pd.to_datetime(correct['date'], errors='coerce')

# Plot time series for each unique indicator in 'correct'
indicators = correct['indicator'].unique()
rows = len(indicators)
cols = 1
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 4))
axes = axes.flatten() if rows > 1 else [axes]  # Handle single subplot case

for i, indicator in enumerate(indicators):
    ax = axes[i]
    indicator_data = correct[correct['indicator'] == indicator]
    ax.plot(indicator_data['date'], indicator_data['seen'], label=f'Correct: {indicator}', color='green')
    ax.set_title(f'Correct: {indicator}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Seen (1 = Yes, 0 = No)')
    ax.legend(loc='upper right')
    ax.grid(True)

plt.tight_layout()
plt.show()
```
