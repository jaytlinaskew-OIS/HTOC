```python
import os
import pandas as pd
import json
from collections import defaultdict

# Define the directory path
data_dir = r"Z:\HTOC\Data_Analytics\Data\OpDiv_Predictions\Full Daily Reports"

# List all files in the directory
files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]

# Read all CSV files into a single DataFrame
df_list = [pd.read_csv(os.path.join(data_dir, file)) for file in files]
full_daily_reports = pd.concat(df_list, ignore_index=True)
files = [f for f in files if f != 'Daily Automation Scripts']

# Define the path to the JSON file
json_path = r"C:\Users\jaskew\Documents\project_repository\notebooks\Indicator Profile\Indicator_Dictionary.json"

# Load existing data if the file exists
if os.path.exists(json_path):
    try:
        with open(json_path, "r", encoding="utf-8") as f:
            existing_data = json.load(f)
    except json.JSONDecodeError:
        existing_data = []
else:
    existing_data = []

# Convert existing data to a set of tuples for duplicate checking
# Only include items that are dicts
existing_set = {tuple(item.items()) for item in existing_data if isinstance(item, dict)}

# Convert current DataFrame to list of dicts
new_data = full_daily_reports.to_dict(orient="records")
new_set = {tuple(item.items()) for item in new_data}

# Combine and remove duplicates
combined_set = existing_set | new_set
combined_data = [dict(items) for items in combined_set]
# Organize combined_data by 'Indicator'

indicator_dict = defaultdict(list)
for item in combined_data:
    indicator = item.get('Indicator')
    if indicator is not None:
        indicator_dict[indicator].append(item)

# Convert defaultdict to regular dict for JSON serialization
organized_data = dict(indicator_dict)
combined_data = organized_data
# Save back to JSON
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(combined_data, f, indent=2)
# Print the number of unique indicators
print(f"Number of unique indicators: {len(combined_data)}")
```
